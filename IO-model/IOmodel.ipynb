```python
import pandas as pd
import numpy as np
```

```python
work_sheet = pd.read_excel('.\\data\\CHN_IOtable_2020.xlsx', sheet_name = 'IOdata_q1_q2')
work_sheet
```

```python
# convert df to array
array_intermediateGoods = work_sheet.iloc[:,2:155].values #.iloc[:,] all rows, 2:154 cols,shape(153, 153)
array_finalUse = work_sheet.iloc[:,155:161].values # shape(153, 6)
array_imports = work_sheet.iloc[:,161].values      # shape(153,1)
array_totalOutput = work_sheet.iloc[:,162].values  # shape(153,1)
#direct consumption coefficients
a_ij_matrix = array_intermediateGoods / array_totalOutput
array_finalUse
```

```python
# 2 calculate complete consumption coefficient
# 2.1 create a (153, 153) identity matrix 
I = np.eye(153)
# 2.2 leontief matrix
leontief = I - a_ij_matrix
# 2.3 inversed leotief matrix (I - a_ij)^-1
leontief_minus = np.linalg.solve(leontief,I)   
#leontief_inverse = np.linalg.inv(leontief)
#leontief_pinv = np.linalg.pinv(leontief)
# 2.4 complete consumption coefficient matrix (optional)
b_ij_matrix = leontief_minus - I

leontief_minus
```

```python
# an endogenous output model: output = (finaluse - import) * (I - aij)^-1
endo_output = np.sum(leontief_minus * (np.sum(array_finalUse, axis = 1) - array_imports), axis = 1)
endo_output
```

```python
# a change in final use -->  change in output?
# mathematical expression: delta_output = (I - aij)^-1 * delta_finalUse
# 3. question: increasing the final use of sector 2 ("林产品") by 1 million CNY ---> output
shock_array = np.zeros((153,1)) # init an array with 0s
row_index = 1  # row to replace (sector 2, index starts from 0)
shock_value = 100 
shock_array[row_index] = shock_value
shock_array
```

```python
# result
new_output = np.sum(leontief_minus * ((np.sum(array_finalUse, axis = 1) + shock_array - array_imports)), axis = 1)

new_output - endo_output
```
